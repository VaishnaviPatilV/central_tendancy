```python



# Mean
mean_age = df["Age"].mean()

# Median
median_age = df["Age"].median()

# Mode
mode_age = df["Age"].mode()[0]

print("Mean Age:", mean_age)
print("Median Age:", median_age)
print("Mode Age:", mode_age)
```

```python
mean_fare = df["Fare"].mean()
median_fare = df["Fare"].median()
mode_fare = df["Fare"].mode()[0]

print("Mean Fare:", mean_fare)
print("Median Fare:", median_fare)
print("Mode Fare:", mode_fare)
```

```python
mean_Parch = df["Parch"].mean()
median_Parch = df["Parch"].median()
mode_Parch = df["Parch"].mode()[0]

print("Mean Parch:", mean_Parch)
print("Median Parch:", median_Parch)
print("Mode Parch:", mode_Parch)
```

```python
mean_SibSp = df["SibSp"].mean()
median_SibSp = df["SibSp"].median()
mode_SibSp = df["SibSp"].mode()[0]

print("Mean SibSp:", mean_SibSp)
print("Median SibSp:", median_SibSp)
print("Mode SibSp:", mode_SibSp)
```
